```python
import pandas as pd

# Example data
y_actual = [1, 0, 0, 1, 0, 1, 0, 0, 1, 0, 1, 0]
y_predicted = [1, 1, 0, 1, 0, 1, 1, 0, 1, 0, 0, 0]

# Creating a DataFrame from the actual and predicted values
df = pd.DataFrame({'y_actual': y_actual, 'y_predicted': y_predicted})

# Creating the confusion matrix
confusion_matrix = pd.crosstab(df['y_actual'], df['y_predicted'], rownames=['Actual'], colnames=['Predicted'])

# Displaying the confusion matrix
print(confusion_matrix)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import ConfusionMatrixDisplay

# Assuming 'cm' is your confusion matrix and 'classes' is the list of class names
cm = np.array([[13, 1], [3, 9]])
classes = ['Class 0', 'Class 1']

# Plotting the confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=classes)

# Customize the plot here (e.g., color map, normalization, etc.)
disp.plot(cmap=plt.cm.Blues)
plt.show()
```
